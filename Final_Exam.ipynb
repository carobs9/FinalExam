```python
from bs4 import BeautifulSoup # importing BeautifulSoup for scraping
import requests as rq
import pandas as pd
```

First, I want to scrape data from the World Cup 2022:

```python
wc_22 = rq.get('https://www.thesoccerworldcups.com/world_cups/2022_top_scorers.php')
wc_22.encoding = 'UTF-8'
```

```python
wc_22.status_code
```

```python
wc_22_html = wc_22.text
wc_22_soup = BeautifulSoup(wc_22_html)
```

Here, I build a list of the top scorers of the 2022 World Cup in order, from those who scored most goals to those who scored less goals:

```python
scorers = wc_22_soup.find_all(class_ ='a-top') # Clean spaces and ugly text
scorers
scorers_list_22 = []
for i in scorers:
    scorers_list_22.append(i.find('a').text)
#scorers_list_22
```

```python
goals_scorers = wc_22_soup.find_all(class_ ='a-top') # Clean spaces and ugly text
goals_scorers
goals_list_22 = []
for i in goals_scorers:
    goals_list_22.append(i.find('b').text)
goals_list_22
```

Now, I scrape data from World Cup 2022:

```python
wc_18 = rq.get('https://www.thesoccerworldcups.com/world_cups/2018_top_scorers.php')
wc_18.encoding = 'UTF-8'
wc_18_html = wc_18.text
wc_18_soup = BeautifulSoup(wc_18_html)
```

```python
scorers_18 = wc_18_soup.find_all(class_ ='a-top') # Clean spaces and ugly text
scorers_18
scorers_list_18 = []
for i in scorers_18:
    scorers_list_18.append(i.find('a').text)
#scorers_list_18
```

```python
goals_scorers_18 = wc_18_soup.find_all(class_ ='a-top') # Clean spaces and ugly text
goals_scorers_18
goals_list_18 = []
for i in goals_scorers_18:
    goals_list_18.append(i.find('b').text)
#goals_list_18
```

Here, I am building an initial dataframe with the top scorers and their respective goals:

```python
# How to make a dataframe out of scraped data:
football_df = pd.DataFrame(list(zip(scorers_list_22, goals_list_22, scorers_list_18, goals_list_18)), columns =['Top Scorers 2022', 'Goals', 'Top Scorers 2018', 'Goals'])
football_df
```
